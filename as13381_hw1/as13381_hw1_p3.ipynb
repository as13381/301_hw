## 3.2

**Answer:**
Retrieval Algorithm 1:

$\begin{align*}
  \mathrm{Accuracy} &= \frac{tp + tn}{tp + tn + fp + fn} = \frac{55 + 25}{55 + 25 + 5 + 15} = 0.8 \\
  \mathrm{Balanced Accuracy} &= \frac{1}{2} (\frac{tp}{tp + fn} + \frac{tn}{tn + fp}) = \frac{1}{2}(\frac{55}{55 + 15} + \frac{25}{25 + 5}) = 0.8095 \\
  \mathrm{F1 score} &= \frac{tp}{tp + \frac{1}{2} (fp + fn)} = \frac{55}{55 + \frac{1}{2} (5 + 15)} = 0.8461
\end{align*}$

Retrieval Algorithm 2:

$\begin{align*}
  \mathrm{Accuracy} &= \frac{tp + tn}{tp + tn + fp + fn} = \frac{60 + 20}{60 + 20 + 10 + 10} = 0.8\\
  \mathrm{Balanced Accuracy} &= \frac{1}{2} (\frac{tp}{tp + fn} + \frac{tn}{tn + fp}) = \frac{1}{2}(\frac{60}{60 + 10} + \frac{20}{20 + 10}) = 0.7619\\
  \mathrm{F1 score} &= \frac{tp}{tp + \frac{1}{2} (fp + fn)} = \frac{60}{60 + \frac{1}{2} (10 + 10)} = 0.8571
\end{align*}$

The instructor is right as F-1 score for the negative class is a better performance metric than balanced accuracy. Since F-1 score is the harmonic mean of precision and recall, it is more suited for evaluating performance on a specific class as it places more importance on true positive predictions and doesn't take into account true negative ones (here instead of using the positive class/relevant as "true positive" we use the negative class/irrelevant). On the other hand, balanced accuracy is the arithmetic mean of sensitivity and specificity, so it is not biased towards any specific class and takes into account both error types. Since we care most about the performance on specifically the negative class, F-1 score for the negative class (which in this case is also the majority class) is the better option. If we cared more/equally about performance on the positive class as well, balanced accuracy would be the better metric.
